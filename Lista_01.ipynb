```python
import numpy as np
import matplotlib.pyplot as plt

#A = np.random.randint(255, size=(800,800))  # Montar uma matriz com valores randomicos entre 0 - 255, de tamanho de 800x800 (EXERCÍCIO 1)
A = np.random.randn(800,800)                 # Montar uma matriz randômica com o mesmo tamanho porém sem valores limitados   
A = A.astype(np.float32)                     # Transformar os dados de A para float32 (EXERCÍCIO 1)
print(A)

print("- Valor Médio:", np.mean(A))           # Valor médio da matriz (EXERCÍCIO 2)
print("- Variância:", np.var(A))              # Variância  da matriz (EXERCÍCIO 2)
print("- Valor Máx:", np.max(A))              # Valor máx da matriz (EXERCÍCIO 2)
print("- Valor Mín:", np.min(A))              # Valor mín (EXERCÍCIO 2)

print("=-=-=Matriz Ruído=-=-=")                       
MR = A[500:763, 500:764]                    # Criou uma matriz ruído, retirado da matria original(A) de tamanho 263x264 A[X, Y] (EXERCÍCIO 3)
MR = np.multiply(MR, 25)                    # Multiplicando a nova matriz por 25 (EXERCÍCIO 3)
plt.imshow(MR, cmap='Dark2')                # Pegando a matriz MR e transformando em imagem em escala dark2 (MATRIZ, em escala dark2) (EXERCÍCIO 3)
plt.show()                                  # Motrar imagem (EXERCÍCIO 3)
print("Imagem da Rosa:")                    
IMG1 = plt.imread('rose_gray.tif')          # Atribuida a variavel com a foto rose_gray (EXERCÍCIO 3)
plt.imshow(IMG1, cmap='Dark2')              # Saida com a imagem da variável, em escala dark2 (EXERCÍCIO 3)
plt.show()

print("Soma das matrizes: (MR + Rose_gray)")
RES = MR+IMG1                                # Variavel Resultado (RES) é a soma de duas matrizes (MR + IMG1) (EXERCÍCIO 3)
plt.imshow(RES, cmap='Dark2')                # Saida com a imagem da variável (SOMA), em escala dark2 (EXERCÍCIO 3)
plt.show()

P = RES[132,263]                        # Valor da matriz no ponto (132,263) atribuida na variável P (EXERCÍCIO 4)
L = IMG1[132,263]                       # Valor da matriz no ponto (132,263) atribuida na variável L (EXERCÍCIO 4)
print("O pixel ruidoso [132,264] tem o valor de:", P)     # Saida do texto do valor encontrado em P (EXERCÍCIO 4)
print("O pixel sem ruido [132,264] tem o valor de:", L)   # Saida do texto do valor encontrado em L (EXERCÍCIO 4)
```
